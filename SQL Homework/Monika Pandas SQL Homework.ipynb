```python
#Create a bar chart of average salary by title.

avg_sal_title_df = pd.read_sql_query('SELECT * FROM avg_salary_by_title', connect)
avg_sal_title_df
```

```python
#Create a bar chart of average salary by title using pandas.

avg_sal_title_df.plot(kind='bar', x='title', y='avg_salary', title='Average Salary by Title')

# Set textual properties
plt.title('Average Salary by Title')
plt.ylabel("AVG Salary")
plt.xlabel("Job Title")

# Show plot
plt.savefig("avg_salary_by_title_bar_chart.png", index=False, header=True)

#plt.show(10)
#plt.tight_layout()
```
